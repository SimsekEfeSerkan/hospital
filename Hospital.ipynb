```python
class hospital:
    def __init__(self, name, surname , age,):
        self.name = name
        self.surname = surname
        self.age = age
    def save(name, surname, age):
        open('hospital.txt', 'a').write(name + ' ' + surname + ' ' + age + '\n')
    def read():
        with open('hospital.txt', 'r') as f:
            for line in f:
                print(line)
    def delete(name, surname, age):
        with open('hospital.txt', 'r') as f:
            lines = f.readlines()
        with open('hospital.txt', 'w') as f:
            for line in lines:
                if line.strip('\n') != name + ' ' + surname + ' ' + age:
                    f.write(line)
    def update(name, surname, age):
        with open('hospital.txt', 'r') as f:
            lines = f.readlines()
        with open('hospital.txt', 'w') as f:
            for line in lines:
                if line.strip('\n') == name + ' ' + surname + ' ' + age:
                    f.write(name + ' ' + surname + ' ' + age + '\n')
                else:
                    f.write(line)
    def search(name, surname, age):
        with open('hospital.txt', 'r') as f:
            lines = f.readlines()
        for line in lines:
            if line.strip('\n') == name + ' ' + surname + ' ' + age:
                print(line)
            else:
                print('Not found')

x = input("what do you want to do?\n1.save\n2.read\n3.delete\n4.update\n5.search\n")
if x == '1':
    name = input("name: ")
    surname = input("surname: ")
    age = input("age: ")
    hospital.save(name, surname, age)
elif x == '2':
    hospital.read()
elif x == '3':  # delete
    name = input("name: ")
    surname = input("surname: ")
    age = input("age: ")
    hospital.delete(name, surname, age)
elif x == '4':  # update 
    name = input("name: ")
    surname = input("surname: ")
    age = input("age: ")
    hospital.update(name, surname, age)
elif x == '5':  # search
    name = input("name: ")
    surname = input("surname: ")
    age = input("age: ")
    hospital.search(name, surname, age)
else:
    print("wrong input")
    exit()
```
